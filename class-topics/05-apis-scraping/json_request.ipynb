```python
import requests
import pandas as pd
```

```python
def get_all_launches():
    
    all_launches_url = "https://api.spacexdata.com/v3/launches"
    
    params = {
        "start": "2017-01-01",
        "end": "2018-01-01"
    }
        
    r = requests.get(all_launches_url, params=params)
    return r.json()

def get_mission_desc(mission_id):
    
    url = f"https://api.spacexdata.com/v3/missions/{mission_id}"
    
    r = requests.get(url)
    
    json_ = r.json()
    
    return json_.get("description")
```

```python
launches = get_all_launches()
```

```python
data = []

for launch in launches:
    
    mission_id_list = launch['mission_id']
    
    try:
        first_id = mission_id_list[0]
    except:
        first_id = None
        
    mission_name = launch["mission_name"]
    launch_date_utc = launch["launch_date_utc"]
    rocket_name = launch["rocket"]["rocket_name"]
    mission_desc = get_mission_desc(first_id)
    
    data.append({
        "mission_name": mission_name,
        "launch_date_utc": launch_date_utc,
        "rocket_name": rocket_name,
        "first_id":first_id,
        "mission_desc": mission_desc
    }
    )
    
print(data)
```
